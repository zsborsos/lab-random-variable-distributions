#### 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
log_transform = data_numerical.copy()
transform = lambda x: np.log(x) if np.isfinite(x) and x!=0 else np.NAN
for col in log_transform:
    log_transform[col] = log_transform[col].apply(transform).fillna(log_transform[col].mean())
```

```python
fig, ax = plt.subplots(2,4, figsize = (14,10))
sns.histplot(log_transform['customer_lifetime_value'], ax = ax[0, 0]) 
sns.histplot(log_transform['income'], ax = ax[0, 1]) 
sns.histplot(log_transform['monthly_premium_auto'], ax = ax[0, 2]) 
sns.histplot(log_transform['months_since_last_claim'], ax = ax[0, 3]) 
sns.histplot(log_transform['months_since_policy_inception'], ax = ax[1, 0]) 
sns.histplot(log_transform['number_of_open_complaints'], ax = ax[1, 1]) 
sns.histplot(log_transform['number_of_policies'], ax = ax[1, 2]) 
sns.histplot(log_transform['total_claim_amount'], ax = ax[1, 3])
plt.tight_layout()
```
